```python
import re
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path
file_path = "./src/logs/2025-03-07_16-13-41_log.txt"

# Read the file content
with open(file_path, "r") as file:
    log_content = file.readlines()

# Extract times for "Ã©tape" and "affichage"
step_times = []
display_times = []
iteration_count = 0

for line in log_content:
    step_match = re.search(r"Temps pour une Ã©tape : ([\d.]+) secondes", line)
    display_match = re.search(r"Temps pour affichage : ([\d.]+) secondes", line)
    
    if step_match:
        step_times.append(float(step_match.group(1)))
        iteration_count += 1
    if display_match:
        display_times.append(float(display_match.group(1)))

# Compute average times
average_step_time = sum(step_times) / len(step_times) if step_times else 0
average_display_time = sum(display_times) / len(display_times) if display_times else 0

# Print average times
print(f"Average time for Ã©tape: {average_step_time:.6f} seconds")
print(f"Average time for affichage: {average_display_time:.6f} seconds")

# Ensure both lists have the same length by truncating to the shorter one
min_length = min(len(step_times), len(display_times))
step_times = step_times[:min_length]
display_times = display_times[:min_length]

# Create a DataFrame for visualization
df = pd.DataFrame({
    "Iteration": range(1, min_length + 1),
    "Step Time (s)": step_times,
    "Display Time (s)": display_times
})

# Plot the step times
plt.figure(figsize=(10, 5))
plt.plot(df["Iteration"], df["Step Time (s)"], label="Temps pour une Ã©tape", marker="o")
plt.xlabel("Iteration")
plt.ylabel("Time (s)")
plt.title("Step Time per Iteration")
plt.legend()
plt.grid(True)
plt.show()

# Plot the display times
plt.figure(figsize=(10, 5))
plt.plot(df["Iteration"], df["Display Time (s)"], label="Temps pour affichage", marker="o", color="r")
plt.xlabel("Iteration")
plt.ylabel("Time (s)")
plt.title("Display Time per Iteration")
plt.legend()
plt.grid(True)
plt.show()
```
